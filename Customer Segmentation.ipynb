## Market / Customer Segmentation
<p>The separation of a market into discrete customer groups with comparable characteristics is known as customer segmentation. Customer segmentation is a useful tool for identifying unmet customer demands. Companies can then outperform the competition by providing products and services that are uniquely enticing.

The most common methods are:

1. Demographic information, such as gender, age, familial and marital status, income, education, and occupation, are the most typical methods in which firms segment their consumer base.

2. Geographical information that varies depending on the company's scope. This information may refer to specific towns or counties for localised enterprises. It could be a customer's location, state, or even nation of residency for larger companies.

3. Psychographics, such as social class, lifestyle, and personality traits.
4. Behavioral data, such as spending and consumption habits, product/service usage, and desired benefits.</p>

## 3. Plotting WCSS & The Elbow Method
<p> Next, I calculated the optimal number of clusters by plotting Within Cluster Sum Of Squares (WCSS) versus the number of clusters (K Value). WCSS calculates the sum of observations' distances from their cluster centroids using the formula below.</p>

<img src="https://miro.medium.com/max/602/0*_3RAyFi3C2zJ-ShA.png">

<b>The Elbow Method</B>
<p>Calculate the Within Cluster Sum of Squared Errors (WSS) for various k values and select the k for which WSS begins to decline. This is seen as an elbow in the WSS-versus-k figure.</p>

The following steps can be summarised as follows:
1. By adjusting K from 1 to 10 clusters, compute K-Means clustering for various values of K.
2. Calculate the total within-cluster sum of squares for each K. (WCSS).
3. Plot the WCSS curve vs the K number of clusters.
4. In most cases, the location of a bend (knee) in the plot is used to determine the proper number of clusters.


```
from sklearn.cluster import KMeans
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(df.iloc[:,1:])
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))    
plt.grid()
plt.plot(range(1,11),wcss, linewidth=2, color="blue", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("WCSS")
plt.show()
```

## 4. Plotting Cluster
<p>Using the elbow technique, the optimal K value is discovered to be 5.</p>

<p>Finally, I created a 3D plot to illustrate the clients' spending score in relation to their annual income. As seen in the 3D graphic, the data points are divided into five classes, each of which is represented by a different colour.</p>

  


```
km = KMeans(n_clusters=5)
clusters = km.fit_predict(df.iloc[:,1:])
df["label"] = clusters

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
# clusters 0-4 (0 indexing)
ax.scatter(df.Age[df.label == 0], df["Annual Income (k$)"][df.label == 0], df["Spending Score (1-100)"][df.label == 0], c='blue', s=60)
ax.scatter(df.Age[df.label == 1], df["Annual Income (k$)"][df.label == 1], df["Spending Score (1-100)"][df.label == 1], c='red', s=60)
ax.scatter(df.Age[df.label == 2], df["Annual Income (k$)"][df.label == 2], df["Spending Score (1-100)"][df.label == 2], c='green', s=60)
ax.scatter(df.Age[df.label == 3], df["Annual Income (k$)"][df.label == 3], df["Spending Score (1-100)"][df.label == 3], c='orange', s=60)
ax.scatter(df.Age[df.label == 4], df["Annual Income (k$)"][df.label == 4], df["Spending Score (1-100)"][df.label == 4], c='purple', s=60)
ax.view_init(30, 185)
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
ax.set_zlabel('Spending Score (1-100)')
plt.show()
```
